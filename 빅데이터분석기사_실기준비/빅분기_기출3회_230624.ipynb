# 작업2유형 

종속 변수 : TravelInsurance , TravelInsurance가 1일 확률을 구해서 제출하라. 평가지표 : auc
제출 파일의 컬럼은 ID, proba 두개만 존재해야한다.

```python
import pandas as pd
train = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e3_p2_train_.csv')
test = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e3_p2_test_.csv')
```

```python
train
```

```python
y = train['TravelInsurance']
```

```python
train = train.drop(columns=['ID','TravelInsurance'])
test_ID = test.pop('ID')
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()

train['Employment Type'] = encoder.fit_transform(train['Employment Type'])
test['Employment Type'] = encoder.fit_transform(test['Employment Type'])

train['GraduateOrNot'] = encoder.fit_transform(train['GraduateOrNot'])
test['GraduateOrNot'] = encoder.fit_transform(test['GraduateOrNot'])

train['FrequentFlyer'] = encoder.fit_transform(train['FrequentFlyer'])
test['FrequentFlyer'] = encoder.fit_transform(test['FrequentFlyer'])

train['EverTravelledAbroad'] = encoder.fit_transform(train['EverTravelledAbroad'])
test['EverTravelledAbroad'] = encoder.fit_transform(test['EverTravelledAbroad'])
```

```python
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score
```

```python
x_train, x_test, y_train, y_test = train_test_split(train,y,test_size=0.2)
```

```python
rf = RandomForestClassifier(max_depth=10,random_state = 100)
rf.fit(x_train,y_train)
rf_pred = rf.predict_proba(x_test)
roc_auc_score(y_test,rf_pred[:,1])
```

```python
xgb = XGBClassifier(max_depth=10,random_state=100,eval_metric='error')
xgb.fit(x_train,y_train)
pred_xgb = xgb.predict_proba(x_test)
roc_auc_score(y_test,pred_xgb[:,1])
```

```python
predict = xgb.predict_proba(test)
```

```python
submission =pd.DataFrame({'ID':test_ID,
                         'proba':predict[:,1]})
submission
```

# 작업3유형

Q1. A,B 공장 각각 정규성을 가지는지 샤피로 검정을 통해 확인하라. (각 공장의 pvalue 출력할 것)

```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e3_p3_2_.csv')
df.head()

from scipy.stats import shapiro

temp = df[(df['group']=='A')]
temp2 = df[(df['group']=='B')]

A = shapiro(temp['rpm'])
B = shapiro(temp2['rpm'])

print(A)
print(B)
```

Q2. A,B 공장 생산 제품의 rpm은 각각 정규성을 가지는지 등분산검정을 통해 확인하라. (각 공장의 pvalue 출력할 것)

```python
from scipy.stats import levene
s , p =levene(temp['rpm'],temp2['rpm'])
round_p = round(p,3)
print(round_p)
```

Q2. 대응 표본 t 검정을 통해 B공장 제품들의 rpm이 A 공장 제품의 rpm보다 크다고 말할 수 있는지 검정하라. pvalue를 소숫점 이하 3자리까지 출력하고 귀무가설, 대립가설 중 하나를 출력하라*

```python
from scipy.stats import ttest_rel
s,p = ttest_rel(temp2['rpm'],temp['rpm'],alternative='greater')
print(round(p,3))
print('대립')
```
