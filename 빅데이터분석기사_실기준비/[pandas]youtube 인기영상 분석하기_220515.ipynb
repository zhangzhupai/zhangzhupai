#### Question 3
채널명을 변경한 채널의 갯수 출력

```python
# channelId가 고유값이므로 channelTitle 갯수와 차이가 있음을 확인 
# channelId value_counts가 2회 이상이면 채널명 변경
print(len(df['channelTitle'].unique()))
print(len(df['channelId'].unique()))

#중복을 제거하고 개수확인
change = df[['channelTitle','channelId']].drop_duplicates().channelId.value_counts()
change.head() # 3회, 2회 변경 기록 확인 가능
print(len(change[change>1])) #1번 초과 변경
```

#### Question 4
일요일에 인기있었던 영상들중 가장많은 영상 종류(categoryId) 출력

```python
#trending_date2 날짜형식으로 변경
df['trending_date2'] = pd.to_datetime(df['trending_date2'])
sunday = df[df['trending_date2'].dt.day_name() == 'Sunday'].categoryId.value_counts().index[0] #첫번째 index
print(sunday)
```

#### Question 5
각 요일별 인기 영상들의 categoryId는 각각 몇개 씩인지 하나의 데이터 프레임으로 표현하라

```python
group = df.groupby([df['trending_date2'].dt.day_name(),'categoryId'],as_index=False).size() #as_index, group하면 그룹 라벨이 인덱스가 되므로 사용하지 않으려면 Flase로 설정 
group.pivot(index='categoryId',columns ='trending_date2')
```
